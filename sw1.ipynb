# 배열의 이용
* 학생의 성적정보를 저장하고, 그 저장된 값에서 과목평균과 학점을 구함.

```python
from IPython.display import Image
import os

dir=os.getcwd()
print dir
Image(filename=os.path.join(dir,'uml/uml_array_avg.png'))
```

```python
marks=[80,90,85,95]
sum=0
for i in marks:
    sum+=i
avg=sum/len(marks)
print "평균: %.2f" % avg
```

# 파일 입출력
* 현재 파일을 열어서 한 줄 씩 출력

```python
from IPython.display import Image
import os

dir=os.getcwd()
print dir
Image(filename=os.path.join(dir,'uml/uml_file.png'))
```

```python
file = open("hello.txt","w")
file.write("say Hello\n")
file.write("안녕하세요\n")
file.close()
file=open("hello.txt","r")
i=0
for line in file.readlines():
    i+=1
    print "%2d줄: %s" % (i,line)
file.close()
```
